```python
import qutip as q
from qutip.nonmarkov.dlheom_solver import HSolverDL
import numpy as np
import matplotlib.pyplot as plt
from itertools import permutations, product
from scipy import constants as c

from DNA.relaxation import loc_deph_op, rate_constant_redfield, glob_therm_op, loc_therm_op, get_eq_state
from utils import calc_trace_distance, calc_purity, calc_coherence, calc_ipr_dm, calc_ipr_hamiltonian
```

```python
def Ham(eps, delta):
    return (eps / 2) * q.sigmaz() + (delta / 2) * q.sigmax() 

def plot_elements(result,elements, HEOM = False):
    t_list = result.times
    element_ops={'11': q.fock_dm(2,0), '22': q.fock_dm(2,1), '12': q.destroy(2), '21': q.create(2)}
    for element in elements:
        if HEOM: pop = q.expect( result.states, element_ops[element] )
        else: pop = result.expect[element]
        if element in ['12','21']: pop = pop.imag
        plt.plot(t_list*1e3, pop) # plotting in fs
```

```python
dims, eps, delta = 2, 20, 40
t_end, t_steps = 2.5, 501 # in ps
elements = ['11','22', '12', '21']
H = Ham(eps,delta)

# HEOM
temperature= 300.0
cutoff_freq= 20.0
reorg_energy= 1.0
spectral_density= 'debye'
matsubara_terms = 12
matsubara_coeffs = np.array([76.84741075-20.j, 12.81657679 +0.j, 6.37666847 +0.j, 4.24723143 +0.j, 3.18440609 +0.j, 2.54714829 +0.j, 2.12245315 +0.j, 1.81915748 +0.j, 1.59171278 +0.j, 1.41482533 +0.j, 1.27332318 +0.j, 1.15755333 +0.j])
matsubara_freqs = np.array([ 20. , 246.77894367, 493.55788733, 740.336831 , 987.11577466, 1233.89471833, 1480.67366199, 1727.45260566, 1974.23154932, 2221.01049299, 2467.78943665, 2714.56838032])
bath_cutoff = 3

# works only for Drude-Lorentz spectral density 
temperature = (temperature * 1e-12 * (c.k / c.hbar)) # K to rad/ps
hsolver = HSolverDL( q.Qobj(H), q.sigmaz() ,reorg_energy, temperature, bath_cutoff, matsubara_terms, cutoff_freq, planck=1.0, boltzmann=1.0, renorm=True, stats=False)

if matsubara_coeffs is not None:
    hsolver.exp_coeff = matsubara_coeffs
if matsubara_freqs is not None:
    hsolver.exp_freq = matsubara_freqs
    
t_list = np.linspace(0, t_end, t_steps) # fs
rho0 = q.fock_dm(2,0)
result = hsolver.run( rho0, t_list)

plot_elements(result, elements, HEOM = True)

eq_state = get_eq_state(H, temperature = temperature)
plt.axhline(y=eq_state[0,0])
plt.axhline(y=eq_state[1,1])
plt.show()
```

```python
# local dephasing Lindblad 

dims, eps, delta = 2, 20, 40
dephasing_rate = 7 # rad/ps
t_end, t_steps = 2.5, 501 # in ps
elements = ['11','22', '12', '21']
H = Ham(eps,delta)

c_ops = []
for site_j in range(dims):
    lind_rate = dephasing_rate
    lind_op = loc_deph_op(dims, site_j)
    c_ops.append( np.sqrt(lind_rate)*lind_op )
    
element_ops={'11': q.fock_dm(2,0), '22': q.fock_dm(2,1), '12': q.destroy(2), '21': q.create(2)}
e_ops = dict( [(element, element_ops[element]) for element in elements ] )

t_list = np.linspace(0, t_end, t_steps) # ps
rho0 = q.fock_dm(2,0)
result = q.mesolve(H, rho0, t_list, c_ops, e_ops )

plot_elements(result, elements)

eq_state = get_eq_state(H, temperature = None)
plt.axhline(y=eq_state[0,0])
plt.axhline(y=eq_state[1,1])
plt.show()
```

```python
# global thermalising Lindblad 
deph_rate= 7
temperature= 300.0
cutoff_freq= 20.0
reorg_energy= 1.0
spectral_density= 'debye'
exponent=1

dims, eps, delta = 2, 20,40
dephasing_rate = 7
t_end, t_steps = 2.5, 501 # in ps
elements = ['11','22', '12', '21']
H = Ham(eps,delta)

eigv, eigs = np.linalg.eig(H)
c_ops = []
for state_a, state_b in permutations(range(dims), 2):
    omega_a, omega_b = eigv[state_a], eigv[state_b]
    lind_rate = rate_constant_redfield( (omega_a - omega_b), deph_rate, cutoff_freq, reorg_energy, temperature, spectral_density, exponent )
    lind_op = glob_therm_op(eigv, eigs, state_a, state_b)
    c_ops.append( np.sqrt(lind_rate)*lind_op )
    
element_ops={'11': q.fock_dm(2,0), '22': q.fock_dm(2,1), '12': q.destroy(2), '21': q.create(2)}
e_ops = dict( [(element, element_ops[element]) for element in elements ] )

t_list = np.linspace(0, t_end, t_steps) # ps
rho0 = q.fock_dm(2,0)
result = q.mesolve(H, rho0, t_list, c_ops, e_ops )

plot_elements(result, elements)

eq_state = get_eq_state(H, temperature = temperature)
plt.axhline(y=eq_state[0,0])
plt.axhline(y=eq_state[1,1])
plt.show()
```

```python
# local thermalising Lindblad 
deph_rate= 7
temperature= 300.0
cutoff_freq= 20.0
reorg_energy= 1.0
spectral_density= 'debye'
exponent=1

dims, eps, delta = 2, 20, 40
dephasing_rate = 7
t_end, t_steps = 2.5, 501 # in ps
elements = ['11', '22', '12', '21']
H = Ham(eps,delta)

eigv, eigs = np.linalg.eigh(H)
c_ops = []
gaps = eigv.reshape(dims, 1) - eigv # matrix that contains all possible eigenenergy differences
unique = np.unique(gaps.flatten()) 
for unique, site_m in product(unique, range(dims)):
    lind_rate = rate_constant_redfield( unique, deph_rate, cutoff_freq, reorg_energy, temperature, spectral_density, exponent )
    lind_op = loc_therm_op(eigv, eigs, unique, site_m)
    c_ops.append( np.sqrt(lind_rate)*lind_op )

element_ops={'11': q.fock_dm(2,0), '22': q.fock_dm(2,1), '12': q.destroy(2), '21': q.create(2)}
e_ops = dict( [(element, element_ops[element]) for element in elements ] )

t_list = np.linspace(0, t_end, t_steps) # ps
rho0 = q.fock_dm(2,0)
result = q.mesolve(H, rho0, t_list, c_ops, e_ops )

plot_elements(result, elements)

eq_state = get_eq_state(H, temperature = temperature)
plt.axhline(y=eq_state[0,0])
plt.axhline(y=eq_state[1,1])
plt.show()
```
