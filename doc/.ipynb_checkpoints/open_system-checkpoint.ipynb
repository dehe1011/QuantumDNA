# Inverse participation ratio 

```python
# inverse participation ratio (IPR) of each eigenstate
# 1/N (localized, coherences in some basis) < IPR < N (delocalized, coherent), IPR = 1 for maximally mixed state (incoherent in any basis)
from DNA import Hamiltonian
from utils import reduced_dm, calc_ipr_dm
import qutip as q
import numpy as np
import matplotlib.pyplot as plt

DNAstring = 'GTTT'
Ham_kwargs = {'relaxation': False}
H = Hamiltonian(DNAstring, Ham_kwargs = Ham_kwargs)
w,v = np.linalg.eigh(H.matrix)

for particle in ['electron', 'hole', 'exciton']:
    ipr_list = np.zeros(H.N**2)
    for i in range(H.N**2):
        dm = q.Qobj( np.outer( v[:,i], v[:,i] ) )
        reduced_dm_ = reduced_dm(dm, particle, H.relaxation )
        ipr_list[i] = calc_ipr_dm(reduced_dm_.full())
    plt.plot(ipr_list, label = particle)
plt.legend()
plt.ylim(0,H.N)
plt.show()
```

```python
# amount of localized Frenkel exciton in excitonic states (eigenstates)
from DNA import Hamiltonian
from utils import reduced_dm, calc_ipr_dm
import qutip as q
import numpy as np
import matplotlib.pyplot as plt

DNAstring = 'GCG'
Ham_kwargs = {'relaxation': False}
H = Hamiltonian(DNAstring, Ham_kwargs = Ham_kwargs)
w,v = np.linalg.eigh(H.matrix)

particle = 'exciton'
pop_list = np.zeros(H.N**2)
for i in range(H.N**2):
    dm = q.Qobj( np.outer( v[:,i], v[:,i] ) )
    reduced_dm_ = reduced_dm(dm, particle, H.relaxation )
    pop_list[i] = np.sum(np.diag(reduced_dm_)).real
plt.plot(pop_list)
plt.ylim(0,1.02)
plt.show()
```
