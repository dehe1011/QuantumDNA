## plot_eigenspectrum

```python
tb_model = TB_Model('WM', (1,3) )
tb_ham = TB_Ham('GCG', tb_model, Ham_kwargs={'relaxation': True, 'particle': 'exciton', 'unit':'rad/ps'})
me_solver = ME_Solver(tb_ham, tb_model, me_kwargs={})
```

```python
fig, ax = plt.subplots()
plot_eigv(ax, tb_ham)
save_fig(fig, 'eigv_plot')
```

```python
fig, ax = plt.subplots()
plot_eigs(ax, tb_ham, 0)
```

```python
num_rows, num_cols = 3,3
fig, axes = plt.subplots(num_rows, num_cols, figsize=(5*num_cols, 5*num_rows) )
for i, ax in enumerate(axes.flatten()[:num_rows*num_cols]):
    plot_eigs(ax, tb_ham, i)
```
