```python
DNAstring = 'GCGCGC'
tb_model = TB_Model('WM', (1,len(DNAstring) ) )
tb_ham = TB_Ham(DNAstring, tb_model, Ham_kwargs={'relaxation': True, 'particle': 'exciton', 'unit':'rad/ps'})
lindblad_diss = Lindblad_Diss(tb_ham,dissipator_kwargs= {'loc_deph_rate': 0,'glob_deph_rate': 0,'relax_rate': 3,'loc_therm': False,'glob_therm': False})
me_solver = ME_Solver(tb_ham, tb_model, lindblad_diss, me_kwargs={})
```

```python
dna_string = 'GCGCGC'
tb_model = TB_Model('WM', (1,len(DNAstring) ) )
J_unit = 'eV'
J_list = np.linspace(0,0.1,100)
J_list *= get_conversion(J_unit, 'rad/ps')

tb_ham = TB_Ham(dnastring, tb_model, Ham_kwargs={'interaction_param': 0, 'source':'Simserides2014'})
tb_basis = tb_ham.tb_model.tb_basis

pop_list = np.zeros( (len(PARTICLES), len(J_list), len(tb_basis)) )
for particle_idx, particle in enumerate(PARTICLES):
    for J_idx, J in enumerate(J_list):
        tb_ham = TB_Ham(dnastring, tb_model, Ham_kwargs={'interaction_param': J, 'source':'Simserides2014'})
        for tb_site_idx, tb_site in enumerate(tb_basis):
            pop_list[particle_idx][J_idx][tb_site_idx] = fourier_analysis_2P(tb_ham, ('(0, 0)', '(0, 0)'), tb_site, particle)[0]

cumulative_pop_list = [0] * ( len(tb_basis) +1 )
running_pop_list = np.zeros( (len(PARTICLES), len(J_list)) )
cumulative_pop_list[-1] = np.array(running_pop_list)
for tb_basis_idx in range(len(tb_basis)):
    running_pop_list += pop_list[:,:,tb_basis_idx]
    cumulative_pop_list[tb_basis_idx] = np.array(running_pop_list)
    
cumulative_pop_list = np.array(cumulative_pop_list)
```

```python
import seaborn as sns
COLORS_DNA_BASES = dict(zip( DNA_BASES, sns.color_palette('colorblind', n_colors=len(DNA_BASES) )))

def get_frame(ax, dna_sequence):
    ax[0].set_ylabel(r'$P_k^{(\mathrm{acc})}$', fontsize=20)
    for particle_idx, particle in enumerate(PARTICLES):
        ax[particle_idx].plot(J_list, [0]*len(J_list), color='k') # plot the bottom line
        ax[particle_idx].set_xlabel(r'$J\,(\mathrm{meV})$')
        ax[particle_idx].set_title(particle, fontsize=20)
        ax[particle_idx].tick_params(axis='both', which='both')
        ax[particle_idx].set_ylim(0, 1.05)
    
    for dna_base_idx, dna_base in enumerate(dna_sequence):
        ax[0].text(0, dna_base_idx/len(dna_sequence), dna_base, fontsize=25, color=COLORS_DNA_BASES[dna_base], alpha=0.6)

fig,ax = plt.subplots(1,3,figsize=(15,5))
for particle_idx, particle in enumerate(PARTICLES):
    for dna_base_idx, dna_base in enumerate(dna_string):
        ax[particle_idx].plot( J_list, cumulative_pop_list[dna_base_idx][:][particle_idx], color='k')
        ax[particle_idx].fill_between(J_list, cumulative_pop_list[dna_base_idx-1][:][particle_idx], cumulative_pop_list[dna_base_idx][:][particle_idx], color=COLORS_DNA_BASES[dna_base], alpha=0.15)

get_frame(ax, dna_string)
```

```python
def dna_base_counter(dna_base: str, dna_dict: Dict[str, float]):
    base_frequency = []
    dna_base_counter = 0 
    for sequence_num, (dna_sequence, val) in enumerate(dna_dict.items()):
        num_sites_per_strand = len(dna_sequence)
        dna_base_counter += dna_sequence.count(dna_base) 
        base_frequency.append( dna_base_counter / (num_sites_per_strand*(sequence_num+1)) )
    return np.array(base_frequency)

fig,ax=plt.subplots(1,1,figsize=(25,5))

ax.plot( list(lifetime_dict.values())[10:], (dna_base_counter('A', lifetime_dict)+dna_base_counter('T', lifetime_dict))[10:]*100, 'o--' )
ax.plot( list(lifetime_dict.values())[10:], (dna_base_counter('G', lifetime_dict)+dna_base_counter('C', lifetime_dict))[10:]*100, 'o--' )

ax.set_ylabel('Percentage',fontsize=25)
ax.set_ylabel('exciton lifetime (ps)',fontsize=25) 
ax.invert_xaxis()
ax.legend(['A-T','G-C'],prop={'size': 30}, loc='upper right')
ax.axhline(50,linestyle='--',color='black',alpha=0.8,lw=2.2)
plt.xticks(fontsize=20), plt.yticks(fontsize=20)
plt.show()
```

```python
def DNA_base_counter(DNA_base: str, DNA_dict: Dict[str, float]):
    DNAstrings = ''.join(DNA_dict.keys())
    max_val, min_val = int( np.floor( max(DNA_dict.values())*1000 )), int( np.floor( min(DNA_dict.values())*1000 ))
    DNA_bases_per_strand = len(list(DNA_dict.keys())[0])
    base_frequency = []
    for val in range(max_val, min_val, -1):
        cut_idx= (np.where( np.array(list(DNA_dict.values()))*1000 >= val )[0][-1] +1) * DNA_bases_per_strand
        base_counter = DNAstrings[:cut_idx].count(DNA_base) / len( DNAstrings[:cut_idx] )
        base_frequency.append( base_counter )
    return np.array(base_frequency)

plt.plot(DNA_base_counter('A', lifetime_dict)+DNA_base_counter('T', lifetime_dict), '.', markersize=12)
plt.plot(DNA_base_counter('G', lifetime_dict)+DNA_base_counter('C', lifetime_dict), '.', markersize=12 )
```
